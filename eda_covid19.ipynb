```python
deaths = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv")
recovered = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv")

display(deaths.head())
display(recovered.head())
```

```python
def last_by_country(data):
    return data.groupby('Country/Region').sum().iloc[:, -1]

last_by_country(confirmed)
```

```python
informations = [last_by_country(confirmed), last_by_country(deaths), last_by_country(recovered)]
combined = pd.concat(informations, axis = 1)
combined.columns = ['confirmed', 'deaths', 'recovered']
combined
```

```python
sum_up = combined.sum()
letality_rate_1 = sum_up['deaths'] / sum_up['confirmed']
letality_rate_2 = sum_up['deaths'] / (sum_up['recovered'] + sum_up['deaths'])
print(letality_rate_1 * 100)
print(letality_rate_2 * 100)
```

```python
letality_rate_1 = combined['deaths'] / combined['confirmed'] * 100
letality_rate_2 = combined['deaths'] / (combined['recovered'] + combined['deaths']) * 100
combined['letality_rate_1'] = letality_rate_1
combined['letality_rate_2'] = letality_rate_2
combined.head()
```

Nenhum dos dados possui algum valor "recovered" distinto de zero, portanto "letality_rate_2" equivale a 100% em todos os casos

## Calcular a taxa de letalidade considerando uma diferença de 12 dias entre o número de casos confirmados e o número de mortes

```python
def last_by_country_at(data, date):
    return data.groupby('Country/Region').sum()[date]
```

```python
informations = [last_by_country_at(confirmed, '2/8/20'), last_by_country_at(deaths, '2/20/20'), last_by_country_at(recovered, '2/20/20')]
combined_12_delay = pd.concat(informations, axis=1)
combined_12_delay.columns = ["confirmed", "deaths", "recovered"]

sum_up = combined_12_delay.loc["China"]
letality_rate_1 = sum_up["deaths"] / sum_up["confirmed"]
print(letality_rate_1 * 100)
```

## Taxa de incidência e mortalidade

```python
# fonte: https://population.un.org/wpp/Download/Standard/CSV/
# acessado em 04/10/2023
population = pd.read_csv('WPP2022_TotalPopulationBySex.csv')
population['PopTotal'] = population['PopTotal'] * 1000
population.head()
```

```python
population.describe()
```

```python
population_total = population.query('Time == 2019')[['Location', 'PopTotal']].set_index('Location')
population_total
```

```python
combined.join(population_total)
```

```python
population.query("Location.str.contains('Hong Kong') and Time==2019", engine = 'python')
```

```python
def rename_location(location):
    if location == 'Republic of Korea':
        return 'South Korea'
    if location == 'Iran (Islamic Republic of)':
        return 'Iran'
    if location == 'United States of America':
        return 'US'
    if location == 'China, Hong Kong SAR':
        return 'Hong Kong'
    else:
        return location
    
population_total['location_for_who'] = population_total.index.map(rename_location)
population_total.query("location_for_who == 'US'")
```

```python
population_total = population_total.set_index('location_for_who')
population_total.head()
```

```python
combined_expanded = combined.join(population_total).sort_values('confirmed', ascending = False).dropna()
combined_expanded.head()
```

```python
combined_expanded['incidence_ratio'] = combined_expanded['confirmed'] / combined_expanded['PopTotal'] * 100000
combined_expanded['mortality_rate'] = combined_expanded['deaths'] / combined_expanded['PopTotal'] * 100000
combined_expanded.head()
```

## Visualização gráfica do progresso da letalidade ao longo do tempo (Global)

```python
confirmed.sum()[3:-2]
```

```python
recovered.sum()[3:]
```

```python
grouped_per_day = pd.DataFrame([confirmed.sum()[3:-2], recovered.sum()[3:], deaths.sum()[3:]],
                               index = ['confirmed', 'recovered', 'deaths']).T
grouped_per_day
```

```python
grouped_per_day['letality_rate_1'] = grouped_per_day['deaths'] / grouped_per_day['confirmed'] * 100
grouped_per_day['letality_rate_2'] = grouped_per_day['deaths'] / (grouped_per_day['deaths'] + grouped_per_day['recovered']) * 100
grouped_per_day.head()
```

```python
plt.figure(figsize = (10,5))
ax = grouped_per_day['letality_rate_1'].plot()
plt.title('Letalidade (Tipo 1) por casos confirmados no mundo todo')
ax.set_ylabel('Letalidade (%)')
plt.show()
```

```python
plt.figure(figsize = (10,5))
ax = grouped_per_day['letality_rate_2'][:-600].plot()
plt.title('Letalidade (Tipo 2) por casos confirmados no mundo todo')
ax.set_ylabel('Letalidade (%)')
plt.show()
```

As duas taxas de letalidade estão convergindo para o mesmo número.
